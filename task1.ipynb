```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("worldpopulationdata.csv")
```

```python
df.head()
```

## Total Population

```python
total_population=df[df['Series Code']=='SP.POP.TOTL']
total_population
```

## Female Population

```python
female_population=df[(df['Series Code']=='SP.POP.TOTL.FE.IN')]
fps=female_population.sort_values(by='2022',ascending=False)
tf=fps.head(10)
```

## Male Population

```python
male_population=df[(df['Series Code']=='SP.POP.TOTL.MA.IN')]
mps=male_population.sort_values(by='2022',ascending=False)
tm=mps.head(10)
```

```python
total_population_sort = total_population.sort_values(by='2022',ascending=False)   
```

```python
top_ten=total_population_sort.head(10)  #highest population
b_ten=total_population_sort.tail(10)   #lowest popultaion
```

## Total population(2022)

```python
plt.figure(figsize=(15, 6))
plt.subplot(2,2,2)
sns.barplot(x='2022',y='Country Name',data=top_ten)
plt.title("Top 10 Countries with according to Populations (2022)")
plt.xlabel('Total population',fontsize=10)
plt.ylabel('Country',fontsize=10)
plt.show()

plt.figure(figsize=(15, 6))
plt.subplot(2,2,2)
sns.barplot(x='2022',y='Country Name',data=b_ten)
plt.title("Bottom 10 Countries according to Total Populations (2022)")
plt.xlabel('Total population',fontsize=10)
plt.ylabel('Country',fontsize=10)
plt.show()
```

```python
merge_data=pd.merge(mps,fps,on="Country Name",suffixes=("_m","_f")) #marged the male and female population data set
```

```python
merge_data["Total Population"] = merge_data["2022_m"] + merge_data["2022_f"] #added a column called total population that stores the data of total population
```

```python
sorted_data = merge_data.sort_values(by="Total Population", ascending=False)
top_10_countries = sorted_data.head(10) #sortted data
b_10_countries = sorted_data.tail(10)
```

## Male and Female population(2022)

```python
# Create the stacked bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x="Country Name", y="2022_f", data=top_10_countries, color="pink", label="Female Population")
sns.barplot(x="Country Name", y="2022_m", data=top_10_countries, bottom=top_10_countries["2022_f"], color="cyan", label="Male Population")
plt.title("Top 10 Countries with Male and Female Populations (2022)")
plt.xlabel("Country")
plt.ylabel("Population")
plt.legend()
plt.xticks(rotation=45, ha="right")
plt.show()

plt.figure(figsize=(12, 6))
sns.barplot(x="Country Name", y="2022_f", data=b_10_countries, color="pink", label="Female Population")
sns.barplot(x="Country Name", y="2022_m", data=b_10_countries, bottom=b_10_countries["2022_f"], color="cyan", label="Male Population")
plt.title("Bottom 10 Countries with Male and Female Populations (2022)")
plt.xlabel("Country")
plt.ylabel("Population")
plt.legend()
plt.xticks(rotation=45, ha="right")
plt.show()
```
